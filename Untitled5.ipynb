## Bokeh

```python
# Ti·∫øp t·ª•c hi·ªÉn th·ªã bi·ªÉu ƒë·ªì ph√¢n t√°n v√† c·ªôt t∆∞∆°ng t√°c sau khi ƒë√£ load d·ªØ li·ªáu
from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import row, column
from bokeh.models import ColumnDataSource, HoverTool, CheckboxGroup, CustomJS, Slider
from bokeh.transform import jitter
from sklearn.preprocessing import MinMaxScaler
import numpy as np

output_notebook()

# Ti·ªÅn x·ª≠ l√Ω v√† chu·∫©n h√≥a
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[['studytime_scaled', 'G3_scaled']] = scaler.fit_transform(df[['studytime', 'G3']])
df_scaled['studytime_jittered'] = df_scaled['studytime_scaled'] + np.random.uniform(-0.03, 0.03, size=len(df_scaled))

# T√°ch theo gi·ªõi t√≠nh
male_df = df_scaled[df_scaled['sex'] == 'M']
female_df = df_scaled[df_scaled['sex'] == 'F']
source_male = ColumnDataSource(male_df)
source_female = ColumnDataSource(female_df)

# Bi·ªÉu ƒë·ªì c·ªôt: S·ªë h·ªçc sinh theo studytime
studytime_counts = df['studytime'].value_counts().sort_index()
bar_source = ColumnDataSource(data=dict(
    studytime=[str(x) for x in studytime_counts.index],
    counts=list(studytime_counts.values)
))
bar_chart = figure(x_range=[str(x) for x in studytime_counts.index],
                   title="S·ªë h·ªçc sinh theo studytime",
                   x_axis_label='Studytime',
                   y_axis_label='S·ªë h·ªçc sinh',
                   width=300,
                   height=450)
bar_chart.vbar(x='studytime', top='counts', width=0.6, source=bar_source,
               color='skyblue', line_color='black')

# Bi·ªÉu ƒë·ªì ph√¢n t√°n
scatter_plot = figure(width=500, height=450,
                      title='Ph√¢n t√°n (jittered): studytime vs G3',
                      x_axis_label='Studytime (jittered)',
                      y_axis_label='G3 (scaled)',
                      tools='pan,wheel_zoom,box_zoom,reset')

r_m = scatter_plot.scatter(x='studytime_jittered', y='G3_scaled', size=6, source=source_male,
                           color='#FFA07A', alpha=0.6, legend_label='Nam')
r_f = scatter_plot.scatter(x='studytime_jittered', y='G3_scaled', size=6, source=source_female,
                           color='#66C2A5', alpha=0.6, legend_label='N·ªØ')

hover = HoverTool(tooltips=[
    ('Gi·ªõi t√≠nh', '@sex'),
    ('Tu·ªïi', '@age'),
    ('S·ªë ng√†y ngh·ªâ', '@absences'),
    ('ƒêi·ªÉm G3', '@G3'),
    ('Studytime', '@studytime')
])
scatter_plot.add_tools(hover)

# Checkbox ·∫©n/hi·ªán nam/n·ªØ
checkbox = CheckboxGroup(labels=["Nam", "N·ªØ"], active=[0, 1])
checkbox_callback = CustomJS(args=dict(r_m=r_m, r_f=r_f, checkbox=checkbox), code="""
    r_m.visible = checkbox.active.includes(0);
    r_f.visible = checkbox.active.includes(1);
""")
checkbox.js_on_change('active', checkbox_callback)

# Slider l·ªçc studytime
slider = Slider(start=1, end=4, value=1, step=1, title="L·ªçc studytime ‚â•")
slider_callback = CustomJS(args=dict(
    src_m=source_male,
    src_f=source_female,
    orig_m=ColumnDataSource(male_df),
    orig_f=ColumnDataSource(female_df),
    slider=slider
), code="""
    const th = slider.value;
    function filter(src, orig) {
        const data = src.data;
        const original = orig.data;
        for (let key in original) data[key] = [];
        for (let i = 0; i < original['studytime'].length; i++) {
            if (original['studytime'][i] >= th) {
                for (let key in original) {
                    data[key].push(original[key][i]);
                }
            }
        }
        src.change.emit();
    }
    filter(src_m, orig_m);
    filter(src_f, orig_f);
""")
slider.js_on_change('value', slider_callback)

# B·ªë c·ª•c layout
layout = row(bar_chart, column(scatter_plot, checkbox, slider))
show(layout)
```

## **A. T-Test**

```python
from scipy.stats import ttest_ind

# T√≠nh trung b√¨nh s·ªë ng√†y ngh·ªâ
mean_absences = df['absences'].mean()

# T√°ch d·ªØ li·ªáu th√†nh hai nh√≥m: ngh·ªâ nhi·ªÅu v√† ngh·ªâ √≠t
group_high_absences = df[df['absences'] > mean_absences]['G3']
group_low_absences = df[df['absences'] <= mean_absences]['G3']

# Th·ª±c hi·ªán ki·ªÉm ƒë·ªãnh t-test (gi·∫£ ƒë·ªãnh ph∆∞∆°ng sai kh√¥ng b·∫±ng nhau)
t_stat, p_value = ttest_ind(group_high_absences, group_low_absences, equal_var=False)

# K·∫øt lu·∫≠n v·ªõi m·ª©c √Ω nghƒ©a alpha = 0.05
alpha = 0.05
conclusion = "Kh√¥ng b√°c b·ªè H0 - Kh√¥ng c√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ" if p_value > alpha else "B√°c b·ªè H0 - C√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ"

# T·ªïng h·ª£p k·∫øt qu·∫£
{
    "Mean absences": mean_absences,
    "S·ªë h·ªçc sinh ngh·ªâ nhi·ªÅu": len(group_high_absences),
    "S·ªë h·ªçc sinh ngh·ªâ √≠t": len(group_low_absences),
    "Mean G3 - Ngh·ªâ nhi·ªÅu": group_high_absences.mean(),
    "Mean G3 - Ngh·ªâ √≠t": group_low_absences.mean(),
    "t-statistic": t_stat,
    "p-value": p_value,
    "K·∫øt lu·∫≠n": conclusion
}
```

# üìä B√°o c√°o ki·ªÉm ƒë·ªãnh t-test: ·∫¢nh h∆∞·ªüng c·ªßa s·ªë ng√†y ngh·ªâ ƒë·∫øn ƒëi·ªÉm s·ªë G3

## üîç Gi·∫£ thuy·∫øt ƒë·∫∑t ra:

- **Gi·∫£ thuy·∫øt g·ªëc (H‚ÇÄ):** Kh√¥ng c√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ v·ªÅ ƒëi·ªÉm G3 gi·ªØa hai nh√≥m h·ªçc sinh (ngh·ªâ nhi·ªÅu v√† ngh·ªâ √≠t).

## **B. Z-Test**

```python
from scipy.stats import norm

# Trung b√¨nh v√† ƒë·ªô l·ªách chu·∫©n to√†n b·ªô
mean_all = df['G3'].mean()
std_all = df['G3'].std()

# Nh√≥m h·ªçc sinh h·ªçc √≠t (studytime <= 2)
group_low_study = df[df['studytime'] <= 2]['G3']
mean_low = group_low_study.mean()
n = len(group_low_study)

# T√≠nh z-statistic
z_stat = (mean_low - mean_all) / (std_all / np.sqrt(n))
# T√≠nh p-value hai ph√≠a
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# K·∫øt lu·∫≠n v·ªõi alpha = 0.05
alpha = 0.05
conclusion = "Kh√¥ng b√°c b·ªè H0 - Kh√¥ng c√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ" if p_value > alpha else "B√°c b·ªè H0 - C√≥ s·ª± kh√°c bi·ªát ƒë√°ng k·ªÉ"

# K·∫øt qu·∫£
{
    "Mean G3 (to√†n b·ªô)": mean_all,
    "Mean G3 (studytime ‚â§ 2)": mean_low,
    "S·ªë h·ªçc sinh studytime ‚â§ 2": n,
    "Z-statistic": z_stat,
    "p-value": p_value,
    "K·∫øt lu·∫≠n": conclusion
}
```
